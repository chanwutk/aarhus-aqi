## Import packages

````python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
````

## Importing data

````python
data = pd.read_json('/content/drive/MyDrive/aarhus/dataset/dataset_x.json')
data.head()
````

### Data fields:


```
          idx : describtion
           0: Luminosity (lux)
           1: Atmospheric Pressure (hPa) 
           2: rain average (mV)
           3: number of cruise ships that day
           4: the total passengers of the cuise
           5: the total crew members of the cuise
           6: how many hours the cruis stays in town
           7: which day in the week does the cruise arrives (0 is Sunday)
           8: month number of when the cruise arrives (1 is January)
           9: date number of when the cruise arrives  
           10: the time of the day that the cruise arrives (24 hours time) 
           11: the mass concentrations of PM2.5 (in ug/m3)
```

````python
data.shape
````

````python
# generate data stats
data.describe()
````

````python
# overall info including type
data.info()
````

## Data Pre-processing

````python
# check null values
data.isnull().sum()
````

````python
# deleting missing values
data = data.dropna()
data.isnull().sum()
````

## Predictions

````python
# excluding labels
features = data
features = features.drop(features.columns[11], axis=1)
features
````

````python
labels = pd.read_json('/content/drive/MyDrive/aarhus/dataset/dataset_y.json')
labels.head()
````

````python
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size = .3)
````

````python
print(x_test.shape, y_test.shape)
````

````python
print(x_train.shape, y_train.shape)
````

````python
lr = LinearRegression()
lr.fit(x_train, y_train)
````

````python
lr.score(x_test, y_test)
````

````python
y_pred = lr.predict(x_test)
y_pred
````

````python
# the mean square error
print('Mean square error (MSE): %.2f ' % mean_squared_error(y_test, y_pred))
````
